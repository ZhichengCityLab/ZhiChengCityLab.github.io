```python
import os

html_template = """
<div class="item" data-id="{keywords}">
    <div class="inner">
        <a href="{project_url}" target="_blank"> 
            <img src="{image_src}" alt="portfolio">
        </a>
        <h3>{project_name}</h3>
        <p class="school">{school_name}</p>
            
    </div>
</div>

"""
```

```python
html_output = []
```

```python


# 文件路径
project_dir = "program_info/summary_as" 
project_image_dir = "program_info/summary_as/image/" 
project_web_url =  "program_page/"
location = "亚洲"

# This function reads a text file and extracts the required information
def extract_project_info(file_path):
    project_info = {
        'project_name': "",
        'school_name': "",
        'keywords': "",
    }
    with open(file_path, 'r', encoding='utf-8') as file:
        for line in file:
            if '项目名称：' in line:
                project_info['project_name'] = line.split('项目名称：')[1].strip()
            elif '学校：' in line:
                project_info['school_name'] = line.split('学校：')[1].strip()
            elif '关键词：' in line:
                keywords = line.split('关键词：')[1].strip().split('，')
                project_info['keywords'] = ' '.join(keywords)
    return project_info


# Generate the HTML for each project

index =0

for file_name in os.listdir(project_dir):
    if file_name.endswith('.txt'):
        file_path = os.path.join(project_dir, file_name)
        project_info = extract_project_info(file_path)
        
        if index <10:
            image_src = project_image_dir + "0" + str(index) + "/logo.png"
            web_url = project_web_url + location + "_0" + str(index) + ".html"
        else:
            image_src =  project_image_dir + str(index) + "/logo.png"
            web_url = project_web_url + location +"_" + str(index) + ".html"
        
        index += 1
        
        # Fill the HTML template with the project information
        html_output.append(html_template.format(
            keywords=project_info['keywords']+ " " + project_info['project_name'] + " " + project_info['school_name'] + " " + location,
            image_src=image_src,
            project_url= web_url,
            project_name=project_info['project_name'],
            school_name=project_info['school_name']
        ))


# Combine all HTML outputs into one string
final_html_output = "\n".join(html_output)

# final_html_output now contains the HTML code for all projects
# You can print it, save it to a file, or use it as needed
print(final_html_output)

```
