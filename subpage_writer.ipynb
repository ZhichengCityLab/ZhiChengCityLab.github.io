```python
def make_whole_subpage( page_context):
    
    web_template= f"""
    <!DOCTYPE html>
    <html lang="en">
    <head>
        <meta charset='utf-8'>
        <meta http-equiv="X-UA-Compatible" content="IE=edge">
        <meta name="viewport" content="width=device-width, initial-scale=1.0">
        <link rel="stylesheet" href="subpage_style.css">
        <title>硕士项目</title>
    </head>
    <body>
        <section class="portfolio" id="Portfolio">
            <div class="container">
                <div class="row">
                    <div class="section-title text-center">
                        <h1>硕士项目</h1>
                    </div>
                </div>
                {page_context}

            </div>
    </body>
    </html>
    """
    
    return web_template
    
```

```python
html_template = """
            <div class="box">
                <img class="img1" src="{image_logo}" alt="">
            </div>

            <div class="section-title">
                <div class="title2">
                    <h2>{project_name}</h2>
                    <h2> <br /></h2>
                    <h4>{school_name}</h4>
                    <h4>{department_name}</h4>
                    <h2> <br /></h2>
                    <h2> <br /></h2>
                    <p>{keywords}</p>
                    <h4><br /></h4>
                    <h4><br /></h4>
                    <h4>项目概述</h4>
                    <h4><br /></h4>
                    <p>{project_intro}</p>
                    <h4><br /></h4>
                    <h4><br /></h4>
                    <h4>项目时长</h4>
                    <h4><br /></h4>
                    <p>{duration}</p>
                    <h4><br /></h4>
                    <h4><br /></h4>
                    <h4>网页链接</h4>
                    <h4><br /></h4>
                    <p>{project_link}</p>
                </div>
            </div>

            <div class="box2">
                <img class="img1" src="{image_image}" alt="">
            </div>
"""
```

```python
# 文件路径
project_image_dir = "image_as/" 
location = "亚洲"
project_dir = "program_info/summary_as" 

html_output=[]

sub_html_output = ''


# This function reads a text file and extracts the required information
def extract_project_info(file_path):
    project_info = {
        'project_name': "",
        'school_name': "",
        'department_name': "",
        'keywords': "",
        'project_intro': "",
        'duration': "",
        'project_link': "",
    }
    
    with open(file_path, 'r', encoding='utf-8') as file:
        for line in file:
            if '项目名称：' in line:
                project_info['project_name'] = line.split('项目名称：')[1].strip()
            elif '学校：' in line:
                project_info['school_name'] = line.split('学校：')[1].strip()
            elif '院系：' in line:
                project_info['department_name'] = line.split('院系：')[1].strip()
            elif '关键词：' in line:
                keywords = line.split('关键词：')[1].strip().split('，')
                project_info['keywords'] = ' '.join(keywords)
            elif '项目概述：' in line:
                project_info['project_intro'] = line.split('项目概述：')[1].strip()
            elif '项目时长：' in line:
                project_info['duration'] = line.split('项目时长：')[1].strip()
            elif '网页链接：' in line:
                project_info['project_link'] = line.split('网页链接：')[1].strip()
    return project_info


# Generate the HTML for each project

index =0

for file_name in os.listdir(project_dir):
    
    if file_name.endswith('.txt'):
        file_path = os.path.join(project_dir, file_name)
        project_info = extract_project_info(file_path)
        
        if index <10:
            image_logo= project_image_dir + "0" + str(index) + "/logo.png"
            image_image= project_image_dir + "0" + str(index) + "/image.png"
            save_path = "program_page/" + location + "_0" + str(index) + ".html"

        else:
            image_logo =  project_image_dir + str(index) + "/logo.png"
            image_image =  project_image_dir + str(index) + "/image.png"
            save_path = "program_page/" +"_" + location + str(index) + ".html"
        
        # Fill the HTML template with the project information
        html_output.append(html_template.format(
            keywords=project_info['keywords'],
            image_logo=image_logo,
            image_image = image_image,
            project_link= project_info['project_link'],
            project_name=project_info['project_name'],
            school_name=project_info['school_name'],
            department_name=project_info['department_name'],
            project_intro=project_info['project_intro'],
            duration=project_info['duration'],
        ))
        
        #print(html_output)
        
        whole_subpage = make_whole_subpage(html_output[index])
        
        print(whole_subpage)
        
        f = open(save_path,'w', encoding = 'utf-8')
        f.write(whole_subpage)
        f.close()
        
        index += 1

        
```
